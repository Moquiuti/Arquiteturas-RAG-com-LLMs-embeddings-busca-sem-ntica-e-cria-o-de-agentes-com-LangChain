```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Base de conhecimento simulada
documentos = [
    "A TRATOTECH oferece garantia de 12 meses para notebooks contra defeitos de fabricação.",
    "O cliente pode cancelar uma compra em até 7 dias após o recebimento do produto.",
    "Produtos com queda, mau uso ou contato com líquido não são cobertos pela garantia.",
    "O suporte técnico funciona de segunda a sexta-feira, das 8h às 18h.",
    "Pagamentos via cartão de crédito podem ser parcelados em até 10 vezes."
]
```

```python
# Carregando um modelo open source de embeddings
modelo = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Convertendo os documentos em embeddings
embeddings_documentos = modelo.encode(documentos)

print("Quantidade de documentos:", len(documentos))
print("Dimensão do vetor de embedding:", embeddings_documentos[0].shape)
```

```python
# Consulta do usuário
consulta = "Meu notebook caiu na água. A garantia cobre?"

# Convertendo a consulta em embedding
embedding_consulta = modelo.encode([consulta])
```

```python
# Calculando similaridade de cosseno entre a consulta e os documentos
similaridades = cosine_similarity(embedding_consulta, embeddings_documentos)[0]

# Ordenando os documentos mais parecidos
indices_ordenados = np.argsort(similaridades)[::-1]

for indice in indices_ordenados:
    print(f"Similaridade: {similaridades[indice]:.4f}")
    print(f"Documento: {documentos[indice]}")
    print("-" * 80)
```
